```python
import pandas as pd #the csv reading library
import matplotlib.pyplot as plt #the ploting library
import glob #to get file names
import numpy as np
from pandas import Series
import seaborn as sns
from matplotlib.pyplot import figure
import cmocean.cm as cmo
import seaborn as sns
```

```python
df_new_3_stats_tc = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Error_test_data/Data_output_error_test_new_method_3_stats_tc.csv')
df_new_2_stats_tc = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Error_test_data/Data_output_error_test_new_method_2_stats_tc.csv')
df_old_stats_tc = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Error_test_data/Data_output_error_test_old_method_stats_tc.csv')
df_new_30_stats_tc = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Error_test_data/Data_output_error_test_new_method_30_stats_tc.csv')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12, 3))
axes.plot(df_new_2_stats_tc['wl'], df_new_2_stats_tc['sem'], color='blue', linewidth=1.5)
axes.plot(df_new_3_stats_tc['wl'], df_new_3_stats_tc['sem'], color='green', linewidth=1.5)
axes.plot(df_new_30_stats_tc['wl'], df_new_30_stats_tc['sem'], color='red', linewidth=1.5)

axes.set_xlabel("Wavelength")
axes.set_ylabel("a (1/m)")
axes.set_xlim(250, 420)
axes.set_ylim(0, 0.0035)
axes.axhline(y=0.00035, color='black', linewidth=1.5)
axes.text(360, .0031, 'Old Method', color='blue')
axes.text(360, .0026, 'New Method', color='green')
axes.text(360, .0021, '30 second Method', color='red')
axes.text(270, .0026, 'line assumes at least Sullivan 2006  35 PSU  ', color='black')
plt.title('Standard Error ')
plt.tight_layout()
#plt.legend()
plt.grid(axis='y')

plt.savefig("../../Dissertation_projects/Absorption_experiments/Figs/Error_test_figs/SEM_temp_corr_summary.png")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12, 3))
axes.plot(df_new_2_stats_tc['wl'], df_new_2_stats_tc['rmse'], color='blue', linewidth=1.5)
axes.plot(df_new_3_stats_tc['wl'], df_new_3_stats_tc['rmse'], color='green', linewidth=1.5)
axes.plot(df_new_30_stats_tc['wl'], df_new_30_stats_tc['rmse'], color='red', linewidth=1.5)

axes.set_xlabel("Wavelength")
axes.set_ylabel("a (1/m)")
axes.set_xlim(250, 420)
axes.set_ylim(0, 0.02)
axes.axhline(y=0.00035, color='black', linewidth=1.5)
axes.text(360, .0175, 'Old Method', color='blue')
axes.text(360, .0155, 'New Method', color='green')
axes.text(360, .0135, '30 second Method', color='red')
axes.text(270, .016, 'line assumes at least Sullivan 2006  35 PSU  ', color='black')
plt.title('Root mean Square Error ')
plt.tight_layout()
#plt.legend()
plt.grid(axis='y')

plt.savefig("../../Dissertation_projects/Absorption_experiments/Figs/Error_test_figs/RMSE_temp_corr_summary.png")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12, 3))
axes.plot(df_new_2_stats_tc['wl'], df_new_2_stats_tc['range'], color='blue', linewidth=1.5)
axes.plot(df_new_3_stats_tc['wl'], df_new_3_stats_tc['range'], color='green', linewidth=1.5)
axes.plot(df_new_30_stats_tc['wl'], df_new_30_stats_tc['range'], color='red', linewidth=1.5)

axes.set_xlabel("Wavelength")
axes.set_ylabel("a (1/m)")
axes.set_xlim(250, 420)
axes.set_ylim(0, 0.08)
axes.axhline(y=0.00035, color='black', linewidth=1.5)
axes.text(360, .07, 'Old Method', color='blue')
axes.text(360, .061, 'New Method', color='green')
axes.text(360, .053, '30 second Method', color='red')
axes.text(270, .065, 'line assumes at least Sullivan 2006  35 PSU  ', color='black')
plt.title('Range')
plt.tight_layout()
#plt.legend()
plt.grid(axis='y')

plt.savefig("../../Dissertation_projects/Absorption_experiments/Figs/Error_test_figs/Range_temp_corr_summary.png")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12, 3))

axes.plot(df_new_3_stats_tc['wl'], df_new_3_stats_tc['mean'], color='red', linewidth=1.5)
axes.plot(df_new_30_stats_tc['wl'], df_new_30_stats_tc['mean'], color='blue', linewidth=1.5)

axes.plot(df_new_3_stats_tc['wl'], df_new_3_stats_tc['sem_up'], color='red', linewidth=.25)
axes.plot(df_new_3_stats_tc['wl'], df_new_3_stats_tc['sem_down'], color='red', linewidth=.25)

axes.plot(df_new_30_stats_tc['wl'], df_new_30_stats_tc['sem_up'], color='blue', linewidth=.25)
axes.plot(df_new_30_stats_tc['wl'], df_new_30_stats_tc['sem_down'], color='blue', linewidth=.25)

axes.fill_between(df_new_3_stats_tc['wl'], df_new_3_stats_tc['sem_up'], df_new_3_stats_tc['sem_down'], alpha=.2, linewidth=0, color='pink')
axes.fill_between(df_new_30_stats_tc['wl'], df_new_30_stats_tc['sem_up'], df_new_30_stats_tc['sem_down'], alpha=.2, linewidth=0, color='lightblue')

axes.set_xlabel("Wavelength")
axes.set_ylabel("a (1/m)")
axes.set_xlim(250, 420)
axes.set_ylim(-0.14, 0.075)
axes.text(260, .025, 'older Method', color='red', size = 20)
axes.text(300, -.09, '30 second method', color='blue', size = 20)
#axs.axhline(y=0.00035, color='black', linewidth=1)
#axs.axhline(y=0.00035, color='black', linewidth=1)
plt.title('Mean and SEM blanks rescan')
plt.tight_layout()
#plt.legend()
plt.grid(axis='y')

plt.savefig("../../Dissertation_projects/Absorption_experiments/Figs/Nitrate_testing/Mean_SEM_methods_30_second.png")
```
