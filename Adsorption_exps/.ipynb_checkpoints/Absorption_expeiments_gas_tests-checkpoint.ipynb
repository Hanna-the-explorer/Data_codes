```python
import pandas as pd #the csv reading library
import matplotlib.pyplot as plt #the ploting library
import glob #to get file names
import numpy as np
from pandas import Series
import seaborn as sns
from matplotlib.pyplot import figure
import cmocean.cm as cmo
import seaborn as sns
```

```python
df1 = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Data_output_Air_pump.csv', sep=',', index_col=0)
df2 = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Data_output_N2_gas.csv', sep=',', index_col=0)
df3 = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Data_output_Air_pump_trash_bag.csv', sep=',', index_col=0)

df_meta = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/gas_meta.csv', sep=',', index_col=0)


df = df2.join(df1)
df = df.join(df3)
df['wl'] = df.index.astype(float)


df_meta
```

```python
# Plotting each column against 'wl'

for column in df.columns:
    if column != 'wl':  # Skip 'wl' column
        plt.plot(df['wl'], df[column], label=column)

plt.xlabel('wl')
plt.ylabel('a')
plt.title('')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
for column in df.columns:
    if column != 'wl':
        color = df_meta.loc[column, 'group_numb'].astype(float)  # Get color from metadata
        color_normalized = (color - 1) / (5.5 - 1)  # Normalize color to range [0, 1]
        color_mapped = cmo.phase(color_normalized)  # Map normalized color to colormap
        axes.plot(df['wl'], df[column], label=column, color=color_mapped)
        
# Add a color bar for the depth on the right side of the figure
fig.subplots_adjust(right=0.85)
cbar_ax = fig.add_axes([0.9, 0.15, 0.03, 0.7])
cbar = fig.colorbar(plt.cm.ScalarMappable(cmap=cmo.phase, norm=plt.Normalize(vmin=0, vmax=5.5)), cax=cbar_ax)
cbar.ax.set_ylabel('Depth', size=15)

plt.xlabel('wl')
plt.ylabel('DO mg/l')
plt.title('')
plt.legend()
plt.grid(True)

#plt.savefig("../Dissertation_projects/Absorption_experiments/Figs/Oxy_exps_1.png")

plt.show()

```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def plot_df_with_groups(df, df_meta, save_name, xlim=None, ylim=None, figsize=(1, 6)):
    # Define the colormap
    unique_groups = df_meta['group'].unique()
    cmap = cm.get_cmap('tab10', len(unique_groups))
    group_colors = {group: cmap(i) for i, group in enumerate(unique_groups)}

    # Create the figure and axis with the specified size
    plt.figure(figsize=figsize)

    # Plot each column in df with the corresponding color from df_meta
    for column in df.columns:
        if column != 'wl':  # Skip 'wl' column
            group = df_meta.loc[column, 'group']
            plt.plot(df['wl'], df[column], label=column, color=group_colors[group])

    # Set the x and y limits if specified
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)

    # Add legend and show the plot
    plt.legend()
    plt.savefig(f"../../Dissertation_projects/Absorption_experiments/Figs/{save_name}.png")
    plt.show()


# Example usage
plot_df_with_groups(df, df_meta,  'Gas_exps_1', xlim=(200, 280), ylim=(-.5, 5), figsize=(10, 8))
plot_df_with_groups(df, df_meta,  'Gas_exps_1', xlim=(200, 230), ylim=(-.5, .5), figsize=(8, 5))


```

```python
import matplotlib.pyplot as plt

def plot_df_with_groups(df, df_meta, group, save_name, xlim=None, ylim=None, figsize=(10, 6)):
    # Create the figure with the specified size
    plt.figure(figsize=figsize)

    # Plot each column in df that belongs to the specified group
    for column in df.columns:
        if column != 'wl' and df_meta.loc[column, 'group'] == group:  # Skip 'wl' column
            plt.plot(df['wl'], df[column], label=column, color=color)

    # Set the x and y limits if specified
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)

    # Add legend and save the plot
    plt.legend()
    plt.savefig(f"../../Dissertation_projects/Absorption_experiments/Figs/{save_name}.png")
    plt.show()

# Example usage
color = 'r'
plot_df_with_groups(df, df_meta, group="N2_flow", save_name="N2_flow", xlim=(200, 230), ylim=(-.4, .4), figsize=(8, 5))
color = 'b'
plot_df_with_groups(df, df_meta, group="air", save_name="air_pump", xlim=(200, 230), ylim=(-.4, .4), figsize=(8, 5))
color = 'g'
plot_df_with_groups(df, df_meta, group="normal", save_name="normal", xlim=(200, 230), ylim=(-.4, 2), figsize=(8, 5))
color = 'black'
plot_df_with_groups(df, df_meta, group="N2_stag", save_name="N2_stagnet", xlim=(200, 230), ylim=(-.4, 5), figsize=(8, 5))
color = 'y'
plot_df_with_groups(df, df_meta, group="trash", save_name="trash", xlim=(200, 230), ylim=(-.4, 1), figsize=(8, 5))

```

```python
# Example usage
def plot_df_with_groups(df, df_meta, group, save_name, xlim=None, ylim=None, figsize=(10, 6)):
    # Create the figure with the specified size
    plt.figure(figsize=figsize)

    # Plot each column in df that belongs to the specified group
    for column in df.columns:
        if column != 'wl' and df_meta.loc[column, 'group'] == group:  # Skip 'wl' column
            plt.plot(df['wl'], df[column], label=column, color=color)

    # Set the x and y limits if specified
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)

    # Add legend and save the plot
    plt.legend()
    #plt.savefig(f"../../Dissertation_projects/Absorption_experiments/Figs/{save_name}.png")
    plt.show()
    
color = 'r'
plot_df_with_groups(df, df_meta, group="N2_flow", save_name="N2_flow", xlim=(200, 230), ylim=(-.4, .4), figsize=(8, 5))
color = 'b'
plot_df_with_groups(df, df_meta, group="air", save_name="air_pump", xlim=(200, 230), ylim=(-.4, .4), figsize=(8, 5))

```

```python
df2 = df
df_Q = df2.drop(['DI_w_N2_gas_stag_1', 'DI_w_N2_gas_stag_2', 'DI_w_N2_gas_stag_3',
                 'DI_w_N2_gas_stag_4', 'DI_w_N2_gas_stag_5',  'DI_open_close_1', 'DI_open_close_2',
                 'Air_pump_1', 'Air_pump_2', 'Air_pump_3', 'Air_pump_4', 'Air_pump_5',
                 'Air_pump_trash_bag_1', 'Air_pump_trash_bag_2', 'Air_pump_trash_bag_3',
                 'Air_pump_trash_bag_4', 'Air_pump_trash_bag_5', 'wl'], axis=1)

df_Q['rmse'] = np.sqrt(((df_Q.sub(df_Q.mean(axis=1), axis=0)) ** 2).mean(axis=1))
df_Q['mean'] = df_Q.mean(axis=1)
df_Q['mean_up'] = df_Q['mean'] + df_Q['rmse']
df_Q['mean_down'] = df_Q['mean'] - df_Q['rmse']
df_Q['wl'] = df_Q.index.astype(float) 
 
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
axes.plot(df_Q['wl'], df_Q['mean'], color = 'black')
axes.plot(df_Q['wl'], df_Q['mean_up'], color = 'grey')
axes.plot(df_Q['wl'], df_Q['mean_down'],  color = 'grey')
        
plt.xlabel('wl')
plt.ylabel('mean and RMSE')
plt.title('N2 gas flow')
plt.legend()
plt.grid(True)

#plt.savefig("../Dissertation_projects/Absorption_experiments/Figs/RMSE_100_ppt_nacl.png")


plt.show()
```
