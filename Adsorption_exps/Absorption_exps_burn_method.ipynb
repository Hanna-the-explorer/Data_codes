```python
import pandas as pd #the csv reading library
import matplotlib.pyplot as plt #the ploting library
import glob #to get file names
import numpy as np
from pandas import Series
import seaborn as sns
from matplotlib.pyplot import figure
import cmocean.cm as cmo
import seaborn as sns
```

```python
df = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Data_output_NaCl_burn_method.csv', sep=',', index_col=0)
df['wl'] = df.index.astype(float)
#df = df.drop('Salt_rep_2', axis=1)


df_meta = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Data_output_Meta_2.csv', sep=',', index_col=0)
```

```python
# Plotting each column against 'wl'
for column in df.columns:
    if column != 'wl':  # Skip 'wl' column
        plt.plot(df['wl'], df[column], label=column)

plt.xlabel('wl')
plt.ylabel('a')
plt.title('')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
for column in df.columns:
    if column != 'wl':
        color = df_meta.loc[column, 'Rep'].astype(float)  # Get color from metadata
        color_normalized = (color - 0) / (5 - 0)  # Normalize color to range [0, 1]
        color_mapped = cmo.phase(color_normalized)  # Map normalized color to colormap
        Group = df_meta.loc[column, 'Group']
        if Group == 'DI':
            linestyle='dashed'
        else:
            linestyle='solid'
        axes.plot(df['wl'], df[column], label=column, color=color_mapped, linestyle=linestyle)
    else: 
        print("")
    

plt.xlabel('wl')
plt.ylabel('Rep')
plt.title('both cuvettes: filterd DI water')



plt.legend()
plt.grid(True)

#plt.savefig("../Dissertation_projects/Absorption_experiments/Figs/test_NaCl_cuv.png")

plt.show()
```

```python
df = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Data_output_NaCl_burn_method.csv', sep=',', index_col=0)
df['wl'] = df.index.astype(float)

import matplotlib.colors as mcolors

df_meta = pd.read_csv('../../Dissertation_projects/Absorption_experiments/Data/Data_output_Meta_2.csv', sep=',', index_col=0)

#made the data frames 
df2 = df
df_di = df2.drop(['NaCl_200_ppt_a', 'NaCl_200_ppt_b', 'NaCl_200_ppt_c',
                  'NaCl_100_ppt_a', 'NaCl_100_ppt_b', 'NaCl_100_ppt_c', 'NaCl_100_ppt_d', 'NaCl_100_ppt_e',
                  #'NaCl_000_ppt_a', 'NaCl_000_ppt_b', 'NaCl_000_ppt_c', 'Blank',
                  'NaCl_035_ppt_a', 'NaCl_035_ppt_b', 'NaCl_035_ppt_c', 
                  'wl'], axis=1)
df_200 = df2.drop([#'NaCl_200_ppt_a', 'NaCl_200_ppt_b', 'NaCl_200_ppt_c',
                  'NaCl_100_ppt_a', 'NaCl_100_ppt_b', 'NaCl_100_ppt_c', 'NaCl_100_ppt_d', 'NaCl_100_ppt_e',
                  'NaCl_000_ppt_a', 'NaCl_000_ppt_b', 'NaCl_000_ppt_c', 'Blank',
                  'NaCl_035_ppt_a', 'NaCl_035_ppt_b', 'NaCl_035_ppt_c', 
                  'wl'], axis=1)
df_100 = df2.drop(['NaCl_200_ppt_a', 'NaCl_200_ppt_b', 'NaCl_200_ppt_c',
                  #'NaCl_100_ppt_a', 'NaCl_100_ppt_b', 'NaCl_100_ppt_c', 'NaCl_100_ppt_d', 'NaCl_100_ppt_e',
                  'NaCl_000_ppt_a', 'NaCl_000_ppt_b', 'NaCl_000_ppt_c', 'Blank',
                  'NaCl_035_ppt_a', 'NaCl_035_ppt_b', 'NaCl_035_ppt_c', 
                  'wl'], axis=1)
df_035 = df2.drop(['NaCl_200_ppt_a', 'NaCl_200_ppt_b', 'NaCl_200_ppt_c',
                  'NaCl_100_ppt_a', 'NaCl_100_ppt_b', 'NaCl_100_ppt_c', 'NaCl_100_ppt_d', 'NaCl_100_ppt_e',
                  'NaCl_000_ppt_a', 'NaCl_000_ppt_b', 'NaCl_000_ppt_c', 'Blank',
                  #'NaCl_035_ppt_a', 'NaCl_035_ppt_b', 'NaCl_035_ppt_c', 
                  'wl'], axis=1)

df_di['rmse'] = np.sqrt(((df_di.sub(df_di.mean(axis=1), axis=0)) ** 2).mean(axis=1))
df_di['mean'] = df_di.mean(axis=1)
df_di['mean_up'] = df_di['mean'] + df_di['rmse']
df_di['mean_down'] = df_di['mean'] - df_di['rmse']
df_di['wl'] = df_di.index.astype(float)

df_200['rmse'] = np.sqrt(((df_200.sub(df_200.mean(axis=1), axis=0)) ** 2).mean(axis=1))
df_200['mean'] = df_200.mean(axis=1)
df_200['mean_up'] = df_200['mean'] + df_200['rmse']
df_200['mean_down'] = df_200['mean'] - df_200['rmse']
df_200['wl'] = df_200.index.astype(float)

df_100['rmse'] = np.sqrt(((df_100.sub(df_100.mean(axis=1), axis=0)) ** 2).mean(axis=1))
df_100['mean'] = df_100.mean(axis=1)
df_100['mean_up'] = df_100['mean'] + df_100['rmse']
df_100['mean_down'] = df_100['mean'] - df_100['rmse']
df_100['wl'] = df_100.index.astype(float)

df_035['rmse'] = np.sqrt(((df_035.sub(df_035.mean(axis=1), axis=0)) ** 2).mean(axis=1))
df_035['mean'] = df_035.mean(axis=1)
df_035['mean_up'] = df_035['mean'] + df_035['rmse']
df_035['mean_down'] = df_035['mean'] - df_035['rmse']
df_035['wl'] = df_035.index.astype(float)

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
axes.plot(df_di['wl'], df_di['mean'], color = 'black')
axes.plot(df_di['wl'], df_di['mean_up'], color = 'grey',  linestyle='dashed')
axes.plot(df_di['wl'], df_di['mean_down'], color = 'grey',  linestyle='dashed')

axes.plot(df_200['wl'], df_200['mean'], color = '#723AC2')
axes.plot(df_200['wl'], df_200['mean_up'], color = '#D5BBFA',  linestyle='dashed')
axes.plot(df_200['wl'], df_200['mean_down'], color = '#D5BBFA', linestyle='dashed')

axes.plot(df_100['wl'], df_100['mean'], color = 'blue')
axes.plot(df_100['wl'], df_100['mean_up'], color = 'lightblue', linestyle='dashed')
axes.plot(df_100['wl'], df_100['mean_down'], color = 'lightblue', linestyle='dashed')

axes.plot(df_035['wl'], df_035['mean'], color = 'green')
axes.plot(df_035['wl'], df_035['mean_up'], color = 'lightgreen', linestyle='dashed')
axes.plot(df_035['wl'], df_035['mean_down'], color = 'lightgreen', linestyle='dashed')

        
plt.xlabel('wl')
plt.ylabel('mean and RMSE')
plt.title('')
plt.xlim(320,430)
plt.ylim(-.15,.4)

#plt.axhline(y=0.00003*100, color="red", linestyle="--")
#plt.axhline(y=-0.00003*100, color="red", linestyle="--")

plt.legend('')
plt.grid(True)

plt.savefig("../../Dissertation_projects/Absorption_experiments/Figs/Burn_method_mean_rmse.png")


plt.show()
```
