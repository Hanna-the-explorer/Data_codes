```python
files = glob.glob('../Dissertation_projects/Data/CTD/DOWNCAST_RENAME/*.csv')
path_save = '../Dissertation_projects/Data/CTD/O2_data_for_MT.csv'

import os
import pandas as pd

# 1. Generate list of files
dir_path = '../Dissertation_projects/Data/CTD/DOWNCAST_RENAME'
all_files = [os.path.join(dir_path, f) for f in os.listdir(dir_path) if f.endswith('.csv')]

# Placeholder for the results
results = []

# 2. Process each file
for file in all_files:
    # Extract metadata from filename
    filename = os.path.basename(file)
    parts = filename.split('_')
    GOM, st, date_part = parts[1], parts[2], parts[3].split('.')[0]
    date = date_part.split('-')

    # Read the file
    df = pd.read_csv(file)

    # Find the row with the maximum 'Pressure [dbar]'
    max_pressure_row = df[df['Pressure [dbar]'] == df['Pressure [dbar]'].max()].iloc[0]

    # Extract required values
    latitude = max_pressure_row['latitude']
    longitude = max_pressure_row['longitude']
    sbeox0Mg_L = max_pressure_row['sbeox0Mg/L']
    sal00 = max_pressure_row['sal00']
    t090C = max_pressure_row['t090C']

    # Save the result
    results.append({
        'GOM': GOM,
        'Station': st,
        'Date': '-'.join(date),
        'Latitude': latitude,
        'Longitude': longitude,
        'O_Mg/L': sbeox0Mg_L,
        'Sal_PSU': sal00,
        'Temp_t090C': t090C
    })

# 3. Convert results to a DataFrame
final_df = pd.DataFrame(results)

# You can save this dataframe to a CSV if required:
# final_df.to_csv('final_results.csv', index=False)
final_df.to_csv(path_save)


final_df
```
