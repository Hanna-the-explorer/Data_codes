```python
files = glob.glob('../Dissertation_projects/Data/CTD/DOWNCAST_RENAME/*.csv')
path_save = '../Dissertation_projects/Data/CTD/Bottom_water_O2_Max_pressure.csv'

import os
import pandas as pd

# 1. Generate list of files
dir_path = '../Dissertation_projects/Data/CTD/DOWNCAST_RENAME'
all_files = [os.path.join(dir_path, f) for f in os.listdir(dir_path) if f.endswith('.csv')]

# Placeholder for the results
results = []

# 2. Process each file
for file in all_files:
    # Extract metadata from filename
    filename = os.path.basename(file)
    parts = filename.split('_')
    GOM, st, date_part = parts[1], parts[2], parts[4].split('.')[0]
    date = date_part.split('-')

    # Read the file
    df = pd.read_csv(file)

    # Find the row with the maximum 'Pressure [dbar]'
    max_pressure_row = df[df['Pressure [dbar]'] == df['Pressure [dbar]'].max()].iloc[0]

    # Extract required values
    latitude = max_pressure_row['latitude']
    longitude = max_pressure_row['longitude']
    sbeox0Mg_L = max_pressure_row['sbeox0Mg/L']
    sal00 = max_pressure_row['sal00']
    t090C = max_pressure_row['t090C']

    # Save the result
    results.append({
        'Trip': GOM,
        'Station': st,
        'Date': '-'.join(date),
        'Latitude': latitude,
        'Longitude': longitude,
        'O_Mg/L': sbeox0Mg_L,
        'Sal_PSU': sal00,
        'Temp_C': t090C
    })

# 3. Convert results to a DataFrame
final_df = pd.DataFrame(results)

# You can save this dataframe to a CSV if required:
# final_df.to_csv('final_results.csv', index=False)
final_df.to_csv(path_save)


final_df
```

```python
import os
import pandas as pd

files = glob.glob('../Dissertation_projects/Data/CTD/ROS_SUMMS/*.csv')
path_save = '../Dissertation_projects/Data/CTD/Niskin_water_O2_Max_depth.csv'

# Placeholder for the results
results = []

# 2. Process each file
for file in all_files:
    # Extract metadata from filename
    filename = os.path.basename(file)
    parts = filename.split('_')

    # Identify and save parts with "G" and "st"
    GOM = next((part for part in parts if 'G' in part), None)
    st = next((part for part in parts if 'St' in part), None)

    # Read the file
    df = pd.read_csv(file)

    # Find the row with the maximum 'Pressure [dbar]'
    max_pressure_row = df[df['Pressure [dbar]'] == df['Pressure [dbar]'].max()].iloc[0]

    # Extract required values
    latitude = max_pressure_row['latitude']
    longitude = max_pressure_row['longitude']
    sbeox0Mg_L = max_pressure_row['sbeox0Mg/L']
    sal00 = max_pressure_row['sal00']
    t090C = max_pressure_row['t090C']

    # Save the result
    results.append({
        'Trip': GOM,
        'Station': st,
        'Latitude': latitude,
        'Longitude': longitude,
        'O_Mg/L': sbeox0Mg_L,
        'Sal_PSU': sal00,
        'Temp_C': t090C
    })

# 3. Convert results to a DataFrame
final_df = pd.DataFrame(results)

# You can save this dataframe to a CSV if required:
# final_df.to_csv('final_results.csv', index=False)
final_df
```

```python
df = pd.read_csv('../Dissertation_projects/Data/meta_data.csv')
df_wc = df[df['Type'] == 'NISK']
df_wc['Depth_M'] = df_wc['Sample_depth'].astype(float)

# Assuming df_wc is your original DataFrame
grouped_idx = df_wc.groupby(['Trip', 'Station'])['Depth_M'].idxmax()
max_depth_df = df_wc.loc[grouped_idx]

max_depth_df
```

```python
# Set "Trip" and "Station" as indices for both dataframes for easier updating
final_df.set_index(['Trip', 'Station'], inplace=True)
max_depth_df.set_index(['Trip', 'Station'], inplace=True)

# Update the 'Depth_M' column in final_df with values from max_depth_df
final_df['Depth_M'] = max_depth_df['Depth_M']

# Reset the index to get back the original structure
final_df.reset_index(inplace=True)

final_df.to_csv(path_save)

final_df
```
