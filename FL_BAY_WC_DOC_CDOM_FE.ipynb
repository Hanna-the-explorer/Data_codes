```python
import pandas as pd #the csv reading library
import matplotlib.pyplot as plt #the ploting library
import glob #to get file names
import numpy as np
from pandas import Series
import seaborn as sns
from matplotlib.pyplot import figure
import cmocean.cm as cmo
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df_doc = pd.read_csv('../FL_bay/DOC_DATA_FLBAY.csv', sep=',', index_col=0)
df_fe = pd.read_csv('../FL_bay/Water_col_iron.csv', sep=',', index_col=0)
df_cdom = pd.read_csv('../FL_bay/CDOM_WC_Outputs.csv', sep=',', index_col=0)
df_cdom.columns
```

```python
df_doc
```

```python
#make all BDL a 0 
df_fe.loc[df_fe['FeT_uM'] == 'BLD', 'FeT_uM'] = 0
df_fe.loc[df_fe['Fe2_uM'] == 'BLD', 'Fe2_uM'] = 0
df_fe.loc[df_fe['FeT_uM'] == 'BDL', 'FeT_uM'] = 0
df_fe.loc[df_fe['Fe2_uM'] == 'BDL', 'Fe2_uM'] = 0

df_fe
```

```python

# Get the common indexes
common_indexes = df_fe.index.intersection(df_doc.index)

# Convert the common indexes to a list
common_indexes_list = common_indexes.tolist()

# Print the list of common indexes
print(common_indexes_list)

import pandas as pd

# Assuming df_iron and df_doc are your DataFrames

# Merge the DataFrames on the common indexes
merged_df = pd.merge(df_fe, df_doc, left_index=True, right_index=True, suffixes=('_fe', '_doc'))

# Print the merged DataFrame
merged_df


cdom_350_join = df_cdom['acdom_350_bc']
df_all = merged_df.join(cdom_350_join)

df_doc = df_doc.join(cdom_350_join)
df_fe = df_fe.join(cdom_350_join)

df_all
```

```python
df_doc
```

```python

x_var = 'NPOC_uM'
y_var = 'acdom_350_bc'

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))

model = LinearRegression()
sub = df_doc.dropna(subset=[x_var, y_var])
data = sub[[x_var, y_var]]
#data = data[data['pw_flux_doc'] < 15]
#data = data[data['pw_flux_tn'] >= 0]
#subcat = data['Camp'].tolist()

X = data[[x_var]]
y = data[y_var]
model.fit(X, y)
score = model.score(X, y)

sns.scatterplot(x=x_var, y=y_var, color='black', s = 100, data=data, ax=axes, legend=False)
sns.lineplot(x=X[x_var], y=model.predict(X), ax=axes, color='black')

axes.set_xlabel(x_var.replace(x_var, 'DOC (\u03BCM)'), size=15)
axes.set_ylabel(y_var.replace(y_var, r'$a_{\mathrm{CDOM}}$ 350 $(m^{-})$  (nm)'), size=15)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
