```python
files = glob.glob('../Dissertation_projects/Data/CDOM/Spectra/sprectra_pchip_baseline_corrected_R*.csv')
path_save = '../Dissertation_projects/Data/CDOM/sprectra_pchip_baseline_corrected_RZ.csv'

def GET_DATA(name):
    df = pd.read_csv(name, sep=',', index_col=0)
    df = pd.DataFrame(df)
    return df

frames = []
for f in files:
    df = GET_DATA(f) # runs the processing function  
    frames.append(df) # adds the new col from each scan to a part of frames 
df = pd.concat(frames, axis=0, ignore_index=False)

df.to_csv(path_save)

df
```
