def contour_plot_tranz_surf_ew(df):
    # Remove NaNs
    df = df.dropna(subset=['longitude', 'Pressure [dbar]', 'sal00'])
    
    # Extract columns
    lon = df['longitude'].values
    depth = df['Pressure [dbar]'].values
    salinity = df[topic].values
    
    # Create grid
    lon_grid = np.linspace(lon.min(), lon.max(), 100)
    depth_grid = np.linspace(depth.min(), 15, 100)
    lon_grid, depth_grid = np.meshgrid(lon_grid, depth_grid)

    # Interpolate salinity data onto the grid, using only valid depths
    salinity_grid = griddata((lon, depth), salinity, (lon_grid, depth_grid), method='linear')

    # Mask out areas below the minimum depth for each longitude
    depth_mask = np.zeros_like(depth_grid, dtype=bool)
    for i in range(len(lon_grid[0])):
        lon_value = lon_grid[0, i]
        valid_depths = depth[lon == lon_value]
        if len(valid_depths) > 0:
            min_depth = valid_depths.min()
            max_depth = valid_depths.max()
            depth_mask[:, i] = (depth_grid[:, i] < min_depth) | (depth_grid[:, i] > max_depth)
        
    salinity_grid = np.ma.masked_where(depth_mask, salinity_grid)

    # Dynamic color range
    #salinity_min = np.nanmin(salinity_grid)
    #salinity_max = np.nanmax(salinity_grid)
    vmin = colormin  # Ensuring vmin is at least 17
    vmax = colormax  # Ensuring vmax is at most 37

    # Create the contour plot
    plt.figure(figsize=(6, 2))
    contour = plt.contourf(lon_grid, depth_grid, salinity_grid, levels=100, cmap=color_pal, vmin=colormin, vmax=colormax)
    plt.gca().invert_yaxis()  # Depths are positive downwards
    plt.xlabel('Longitude')
    plt.ylabel('Depth (m)')
    plt.title(title)
    
    # Show the plot
    plt.tight_layout()
    plt.savefig(f'../Dissertation_projects/Figures/CTD_profiles/Carbon_cyc_paper_CTD_figs/{save_name}.png')
    plt.show()

```python
from scipy.interpolate import griddata

dsx = ds[(ds['camp'] == 'GOM22SP') & (ds['station'].isin(['St.6', 'St.5B', 'St.4', 'St.MK', 'St.ML', 'St.MK-2','St.16', 'St.16-2', 'St.2','7' ]))]

# Extract columns
lon = dsx['longitude'].values
depth = dsx['Pressure [dbar]'].values
salinity = dsx['sal00'].values

# Create grid
lon_grid = np.linspace(lon.min(), lon.max(), 100)
depth_grid = np.linspace(depth.min(), depth.max(), 100)
lon_grid, depth_grid = np.meshgrid(lon_grid, depth_grid)

# Interpolate salinity data onto the grid, using only valid depths
salinity_grid = griddata((lon, depth), salinity, (lon_grid, depth_grid), method='linear')

# Mask out areas below the minimum depth for each longitude
depth_mask = np.zeros_like(depth_grid, dtype=bool)
for i in range(len(lon_grid[0])):
    lon_value = lon_grid[0, i]
    valid_depths = depth[lon == lon_value]
    if len(valid_depths) > 0:
        min_depth = valid_depths.min()
        max_depth = valid_depths.max()
        depth_mask[:, i] = (depth_grid[:, i] < min_depth) | (depth_grid[:, i] > max_depth)
        
salinity_grid = np.ma.masked_where(depth_mask, salinity_grid)

# Create the contour plot
plt.figure(figsize=(12, 4))
contour = plt.contourf(lon_grid, depth_grid, salinity_grid, levels=100, cmap='viridis')
plt.colorbar(contour, label='Salinity')
plt.gca().invert_yaxis()  # Depths are positive downwards
plt.xlabel('longitude')
plt.ylabel('Depth (m)')
plt.title('Salinity Contour Plot')
plt.show()
```
