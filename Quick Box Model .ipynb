```python
import pandas as pd #the csv reading library
import matplotlib.pyplot as plt #the ploting library
import glob #to get file names
import numpy as np
from pandas import Series
import seaborn as sns
from matplotlib.pyplot import figure
import cmocean.cm as cmo
import seaborn as sns
```

```python
flux_doc_pw = pd.read_csv('../Dissertation_projects/Data/Stats_data_sets/Flux_doc_pw.csv', sep=',', index_col=0)

july2116 = 13/1000/12.011*1000*17867.93*86400
july2128 = 8.97/1000/12.011*1000*18547.53*86400
oct2115 = 10.3/1000/12.011*1000*7475.65*86400
nov2112 = 13.01/1000/12.011*1000*11468.32*86400

df_near = flux_doc_pw[flux_doc_pw['Stn'].isin(['St.7', 'St.16', 'St.2', 'St.9', 'St.14'])]
df_all = flux_doc_pw[flux_doc_pw['Stn'].isin(['St.7', 'St.16', 'St.2', 'St.9', 'St.14', 'St.13', 'St.12', 'St.15'])]

mean_all_fall = df_all[df_all['Camp'] == 'GOM21FA']['pw_flux_doc'].mean()
mean_near_fall = df_near[df_near['Camp'] == 'GOM21FA']['pw_flux_doc'].mean()
mean_all_summer = df_all[df_all['Camp'] == 'GOM2021']['pw_flux_doc'].mean()
mean_near_summer = df_near[df_near['Camp'] == 'GOM2021']['pw_flux_doc'].mean()

july21flux_near = mean_near_summer/1000*1000000*4788
july21flux_all = mean_all_summer/1000*1000000*8650
oct21flux_near = mean_near_fall/1000*1000000*4788
oct21flux_all = mean_all_fall/1000*1000000*8650

#final units on this all is mols per day 


print(july2116)
print(july2128)
print(oct2115)
print(nov2112)
print('')
print(july21flux_near)
print(july21flux_all)
print(oct21flux_near)
print(oct21flux_all)
```

```python
# Assuming you have calculated values for july2116, july2128, oct2115, nov2112, 
# july21flux_near, july21flux_all, oct21flux_near, oct21flux_all


# Creating lists for x-axis labels and corresponding values
labels = ['July 16 river', 'July 28 river', 'Oct 15 river', 'Nov 12 river', 'July flux Near', 'July flux All', 'Oct flux Near', 'Oct flux All']
values = [july2116, july2128, oct2115, nov2112, july21flux_near, july21flux_all, oct21flux_near, oct21flux_all]

# Plotting the bar graph
plt.figure(figsize=(10, 3))

plt.bar(labels, values)
plt.ylabel('Mean pw_flux_doc')  # You can adjust the ylabel as needed
plt.title('Mean pw_flux_doc for Different Periods and Stations')
plt.show()
```

```python
wc_doc_stocks = pd.read_csv('../Dissertation_projects/Data/Stats_data_sets/Water_col_DOC_stocks.csv', sep=',', index_col=0)

index_values = ['St.7_GOM21FA', 'St.16_GOM21FA', 'St.2_GOM21FA', 'St.9_GOM21FA', 'St.14_GOM21FA', 'St.12_GOM21FA', 'St.15_GOM21FA']
wc_all_fall = wc_doc_stocks.loc[index_values]

index_values = ['St.7_GOM21FA', 'St.16_GOM21FA', 'St.2_GOM21FA', 'St.9_GOM21FA', 'St.14_GOM21FA']
wc_near_fall = wc_doc_stocks.loc[index_values]

index_values = ['St.7_GOM2021', 'St.16_GOM2021', 'St.2_GOM2021', 'St.9_GOM2021', 'St.14_GOM2021', 'St.13_GOM2021', 'St.15_GOM2021']
wc_all_summer = wc_doc_stocks.loc[index_values]

index_values = ['St.7_GOM2021', 'St.16_GOM2021', 'St.2_GOM2021', 'St.9_GOM2021', 'St.14_GOM2021']
wc_near_summer = wc_doc_stocks.loc[index_values]

wc_mean_all_fall = wc_all_fall['NPOC_uM_per_cubic_meter'].mean()
wc_mean_all_fall_dep = wc_all_fall['max_depth'].mean()

wc_mean_near_fall = wc_near_fall['NPOC_uM_per_cubic_meter'].mean()
wc_mean_near_fall_dep = wc_near_fall['max_depth'].mean()

wc_mean_all_summer = wc_all_summer['NPOC_uM_per_cubic_meter'].mean()
wc_mean_all_summer_dep = wc_all_summer['max_depth'].mean()

wc_mean_near_summer = wc_near_summer['NPOC_uM_per_cubic_meter'].mean()
wc_mean_near_summer_dep = wc_near_summer['max_depth'].mean()

wc_stock_all_fall = wc_mean_all_fall*wc_mean_all_fall_dep*1000000*8650
wc_stock_near_fall = wc_mean_near_fall*wc_mean_near_fall_dep*1000000*4788
wc_stock_all_summer = wc_mean_all_summer*wc_mean_all_summer_dep*1000000*8650
wc_stock_near_summer = wc_mean_near_summer*wc_mean_near_summer_dep*1000000*4788
```
