```python
import pandas as pd #the csv reading library
import matplotlib as mpl
import matplotlib.pyplot as plt #the ploting library
import glob #to get file names
import numpy as np
from pandas import Series
import seaborn as sns
from matplotlib.pyplot import figure
import cmocean.cm as cmo
import seaborn as sns
```

```python
files = glob.glob('../FL_bay/CDOM/OUTPUTS/CDOM_outputs_WC_FLBAY_*.csv')

def GET_DATA(name):
    df = pd.read_csv(name, sep=',', index_col=0)
    df = pd.DataFrame(df)
    return df

frames = []
for f in files:
    df = GET_DATA(f) # runs the processing function  
    frames.append(df) # adds the new col from each scan to a part of frames 
df = pd.concat(frames, axis=0, ignore_index=False)

df
```

```python
df['id'] = df.index
    
# Convert 'id' column to numeric, coercing errors to NaN
df['id'] = pd.to_numeric(df['id'], errors='coerce')

# Drop rows where 'id' column contains NaN (non-numeric values)
df = df.dropna(subset=['id'])

df['id'] = df['id'].astype(object)

df
```

```python
df['Sample_ID'] = df['id'].astype(int).astype(str)

df.index = df['Sample_ID']

df = df.drop('id', axis=1)
#df = df.drop('file_id', axis=1)

df
```
