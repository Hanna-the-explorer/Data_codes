```python
#make function 
def GET_DATA(name):
    df = pd.read_csv(name, sep=',', index_col=0)
    df = pd.DataFrame(df)
    return df
```

```python
#add it all up 
frames = []
for f in files:
    df = GET_DATA(f) # runs the processing function  
    frames.append(df) # adds the new col from each scan to a part of frames 
df = pd.concat(frames, axis=0, ignore_index=False)
df.index = df.index.astype(str)


df["FeT_uM"] = df["FeT Conc (uM)"]
df.index = df.index.astype(str)


df.to_csv('../Fellowships/IRL_GRF/FE/IRL_IRON_FeT.csv')

```

```python
# Start with the original code to combine all the DataFrames
frames = []
for f in files:
    df = GET_DATA(f)  # runs the processing function  
    frames.append(df)  # adds the new col from each scan to a part of frames 

df = pd.concat(frames, axis=0, ignore_index=False)
df.index = df.index.astype(float)

# Drop rows with NaN index values
df = df[df.index.notna()]

df.index = df.index.astype(int).astype(str)

# Drop rows where Notes column has value "X"
df = df[df["Notes"] != "X"]

# Drop the specified columns
columns_to_drop = ["Notes", "Notes_CS", "Notes_BL", "Notes_LD", "Notes_MS"]
df.drop(columns=columns_to_drop, inplace=True)

# Rename columns
df.rename(columns={"FeT Conc (mg/L)": "FeT_mgl", "FeT Conc (uM)": "FeT_uM"}, inplace=True)

# Insert the new column 'type' with a constant value "WC"
df["type"] = "WC"

# Rearrange columns
df = df[[ 'Treatment', 'Campaign', 'Location', 'type', 'Date', 'FeT_uM', 'FeT_mgl', 'Depth']]

df
```
