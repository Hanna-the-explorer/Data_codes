```python
ds = pd.read_csv('../Dissertation_projects/Experiments/STEPWISE_EXPS/STEPWISE_CDOM_FE_DATA.csv', sep=',', index_col=0)
ds = ds[ds['group'].isin(['one','seven', 'eight','nine', 'ten', 'eleven'])]

# Function to normalize a column
def normalize(column):
    return (column - column.min()) / (column.max() - column.min())

# Apply the normalization for each unique group in 'group'
ds[[ 'a350_norm', 'time_norm']] = ds.groupby('group')[[ 'a350', 'Time']].transform(normalize)

ds.loc[ds['FeT_uM'] == 131.8388, 'FeT_uM'] = np.nan
ds.loc[ds['Fe3_uM'] == 119.8194, 'Fe3_uM'] = np.nan


# Initialize the normalized columns in the DataFrame
ds['Fe2_norm'] = None
ds['Fe3_norm'] = None

# Loop over each unique group in 'group'
for group in ds['group'].unique():
    # Subset the data for the current group
    group_data = ds[ds['group'] == group]
    
    # Calculate the min and max values across both 'Fe2_uM' and 'Fe3_uM'
    combined_min = min(group_data['Fe2_uM'].min(), group_data['Fe3_uM'].min())
    combined_max = max(group_data['Fe2_uM'].max(), group_data['Fe3_uM'].max())
    
    # Normalize 'Fe2_uM' and 'Fe3_uM' using the combined min and max
    ds.loc[ds['group'] == group, 'Fe2_norm'] = (group_data['Fe2_uM'] - combined_min) / (combined_max - combined_min)
    ds.loc[ds['group'] == group, 'Fe3_norm'] = (group_data['Fe3_uM'] - combined_min) / (combined_max - combined_min)
```
