```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import xarray as xr
from pathlib import Path
import cartopy
import cartopy.crs as ccrs
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import cartopy.feature as cfeature
import cmocean
import cmocean.cm as cmo
```

```python
plt.rcParams.update({"font.size": 20})
SMALL_SIZE = 22
MEDIUM_SIZE = 22
LARGE_SIZE = 26
plt.rc("font", size=SMALL_SIZE)
plt.rc("xtick", labelsize=SMALL_SIZE)
plt.rc("ytick", labelsize=SMALL_SIZE)
plt.rc("axes", titlesize=SMALL_SIZE)
plt.rc("legend", fontsize=SMALL_SIZE)
```

```python
# Open prepared bathymetry dataset using pathlib to sepcify the relative path

#data from https://data.noaa.gov/metaview/page?xml=NOAA/NESDIS/NGDC/MGG/DEM/iso/xml/1521.xml&view=getDataView&header=none
bathy_file_path = Path('../Dissertation_projects/Data/Bathy/southern_louisiana_13_navd88_2010.nc')
bathy_ds = xr.open_dataset(bathy_file_path)
bathy_lon, bathy_lat, bathy_h = bathy_ds.x, bathy_ds.y, bathy_ds.z.values
bathy_h[bathy_h > 0] = 0
```

```python
bathy_conts = np.arange(bathy_h.min(), 500, 500)
```

```python
lat_top=bathy_lat.values.max()
lat_bot=bathy_lat.values.min()
lon_left=bathy_lon.values.min()
lon_rigth=bathy_lon.values.max()
extent=[lon_left, lon_rigth, lat_bot, lat_top]

coord = ccrs.PlateCarree()
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection=coord)
ax.set_extent(extent, crs=coord);
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection=coord)
ax.set_extent(extent, crs=coord)
bathy = ax.contourf(bathy_lon, bathy_lat, bathy_h, bathy_conts, transform=coord, cmap="Blues_r")
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection=coord)
ax.set_extent(extent, crs=coord)
bathy = ax.contourf(bathy_lon, bathy_lat, bathy_h, transform=coord, cmap=cmo.deep)
```

```python
lat_top=29.7
lat_bot=28
lon_left=-91.5
lon_rigth=-88.75

extent=[lon_left, lon_rigth, lat_bot, lat_top]

res = '10m'

def plot_mapbase(extent, ax=None, res='10m', projection=ccrs.PlateCarree()):
    if ax is None:
        fig = plt.figure(figsize=(10,5)) # set fig size
        ax = fig.add_subplot(111, projection=projection)
        
    ax.set_extent(extent)
    ax.coastlines(res)
    ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                                name='land',
                                                facecolor='lightgrey', 
                                                scale=res)) 
    ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                                name='rivers_lake_centerlines',
                                                edgecolor='blue',
                                                facecolor='none',
                                                scale=res))
    gl = ax.gridlines(xlocs=np.arange(lon_left, lon_rigth+1),
                      ylocs=np.arange(lat_bot, lat_top+1),
                      draw_labels=True, 
                      linewidth=0.5, color='k', linestyle=':')
    gl.top_labels = gl.right_labels = False
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    return ax

fig, axes = plt.subplots(figsize=(20, 20), subplot_kw={'projection': ccrs.PlateCarree()})
bathy = axes.contourf(bathy_lon, bathy_lat, -bathy_h, transform=coord, cmap=cmo.deep)
plot_mapbase(extent, ax=axes)
```
