```python
# slopes Calculate 
Frames = []

def doc_bc_cal(code, time): 
    ds = df_doc[df_doc['code'] == code]
    ds = ds.sort_values(by='Time')

    if code == 'St.5B_GOM22SP':
        ds = ds[ds['dil_fac'] <= 2]
    elif code == 'St.14_GOM22SU':
        ds = ds[ds['Time'] >= 4]
    else:
        pass

    ds_cut = ds[ds['Time'] <= time]
    
    x = ds_cut['Time'].ravel()
    y = ds_cut['NPOC_uM'].ravel()
    colors = ds['dil_fac'].astype(float)

    vmin = 0
    vmax = 5
    
    # Calculate the slope of the regression line
    if len(x) > 1:
        slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    
        # Add the slope value to the plot
        slope = slope
        rsquared = r_value**2
        std_err = std_err
    
        dic = {'code':[code],
               'slope_doc_bc':[slope],
               'rsquared_doc_bc':[rsquared], 
               'std_err_doc_bc':[std_err]}
    
        slopes = pd.DataFrame(dic)
        slopes = slopes.set_index('code', drop=True)

        fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(4, 2)) 
        axs.plot(x, intercept + slope * x, c='black') 
        axs.scatter(ds['Time'], ds['NPOC_uM'], c=colors, cmap=cmo.deep, vmin=vmin, vmax=vmax)
        axs.set_ylabel('DOC uM')
        axs.set_xlabel('time')
        fig.suptitle(code)
    
        Frames.append(slopes)


doc_bc_cal('St.11_GOM22SU', 5)

doc_bc_cal('St.12_GOM21FA', 10)
doc_bc_cal('St.12_GOM22SU', 10)

doc_bc_cal('St.13_GOM21FA', 10)
doc_bc_cal('St.13_GOM22SP', 10)
doc_bc_cal('St.13_GOM22SU', 6)

doc_bc_cal('St.14_GOM21FA', 6)
doc_bc_cal('St.14_GOM22SP', 6)
doc_bc_cal('St.14_GOM22SU', 7)

doc_bc_cal('St.15_GOM21FA', 12)
doc_bc_cal('St.15_GOM22SP', 10)
doc_bc_cal('St.15_GOM22SU', 6)

doc_bc_cal('St.16_GOM22SP', 4.5)
doc_bc_cal('St.16_GOM22SU', 10)

doc_bc_cal('St.2_GOM22SP', 5.5)
doc_bc_cal('St.2_GOM22SU', 3)

doc_bc_cal('St.4_GOM22SP', 4.5)
doc_bc_cal('St.4_GOM22SU', 4.5)

doc_bc_cal('St.5B_GOM21FA', 10)
doc_bc_cal('St.5B_GOM22SP', 6)

doc_bc_cal('St.7_GOM21FA', 10)
doc_bc_cal('St.7_GOM22SP', 12)
doc_bc_cal('St.7_GOM22SU', 7)

doc_bc_cal('St.9_GOM21FA', 7.5)
doc_bc_cal('St.9_GOM22SP', 10)
doc_bc_cal('St.9_GOM22SU', 7.5)

doc_bc_cal('St.MK_GOM22SP', 5)
doc_bc_cal('St.MK_GOM22SU', 5)

############# removed by bad bc high ########### 
#Summer 2021 MK, 5B, 4, 7
#Fall 2021 4, 2, MK
#spring 2022 
#summer 2022 5B

#doc_bc_cal('St.MK_GOM21FA', 10)
#doc_bc_cal('St.5B_GOM22SU', 10)



```
