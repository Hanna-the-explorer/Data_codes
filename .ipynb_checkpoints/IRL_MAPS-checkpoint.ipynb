```python
# Change this!
lat_top=29
lat_bot=27.0
lon_left=-81.5
lon_rigth=-79.7
extent=[lon_left, lon_rigth, lat_bot, lat_top]

# set resolution of the features in map: '10m', '50m', or '110m'.  
# If you do this the first time it will take a while downloading the coastline 
# and other features. In the future it will be faster
res = '10m'

#Figure

fig = plt.figure(figsize=(10,5)) # set fig size. You will probably want to change this
ax = fig.add_subplot(111, projection=ccrs.PlateCarree()) #set cartographic projection. There are a bunch
ax.set_extent(extent) # set the extent, or the box of the map
ax.coastlines(res) # set coastilines, and resolution 

#land
ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                            name='land',
                                            facecolor='grey', # change color here
                                            scale=res)) 
#Rivers
ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                            name='rivers_lake_centerlines',
                                            edgecolor='blue', # change color here
                                            facecolor='none',
                                            scale=res))

#Gir. Only if you want one:
gl = ax.gridlines(xlocs=np.arange(lon_left, lon_rigth+1),
                  ylocs=np.arange(lat_bot, lat_top+1),
                  draw_labels=True, 
                  linewidth=0.5, color='k', linestyle=':')
gl.top_labels = gl.right_labels = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
```

```python
#making it a function

def plot_mapbase(extent, res='10m', projection=ccrs.PlateCarree()):
    fig = plt.figure(figsize=(36,48)) # set fig size
    ax = fig.add_subplot(111, projection=projection)
    ax.set_extent(extent)
    ax.coastlines(res)
    
    #land
    ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                                name='land',
                                                facecolor='#F7FAF7', # change color here
                                                scale=res)) 
    
    #ocean
    ax.add_feature(cfeature.NaturalEarthFeature(category='physical', 
                                                name='ocean',
                                                facecolor='#F7F9FA', # change color here
                                                scale=res)) 
    
    ax.coastlines(resolution='10m', color='black', linestyle='-', linewidth=7.0, alpha=1)
    
    #gl = ax.gridlines(xlocs=np.arange(lon_left, lon_rigth+1),
    #                  ylocs=np.arange(lat_bot, lat_top+1),
    #                  draw_labels=True, 
    #                  linewidth=0.5, color='k', linestyle=':')
    #gl.top_labels = gl.right_labels = False
    #gl.xformatter = LONGITUDE_FORMATTER
    #gl.yformatter = LATITUDE_FORMATTER
    #gl.xlabel_style = {'size': 15, 'color': 'black'}
    #gl.ylabel_style = {'size': 15, 'color': 'black'}
    return fig, ax

fig, ax = plot_mapbase(extent)

plt.savefig('../Fellowships/IRL_GRF/IRL_2023/Plots/IRL_map.png')
```

```python
def plot_mapbase_diff(extent, res='10m', projection=ccrs.PlateCarree()):
    fig = plt.figure(figsize=(30,15)) # set fig size
    ax = fig.add_subplot(111, projection=projection)
    ax.set_extent(extent)
    ax.coastlines(res)
    ax.coastlines(resolution='10m', color='lightgrey', linestyle='-', alpha=1)
    #gl = ax.gridlines(xlocs=np.arange(lon_left, lon_rigth+1),
    #                  ylocs=np.arange(lat_bot, lat_top+1),
    #                  draw_labels=True, 
    #                  linewidth=0.5, color='k', linestyle=':')
    #gl.top_labels = gl.right_labels = False
    #gl.xformatter = LONGITUDE_FORMATTER
    #gl.yformatter = LATITUDE_FORMATTER
    #gl.xlabel_style = {'size': 15, 'color': 'black'}
    #gl.ylabel_style = {'size': 15, 'color': 'black'}
    return fig, ax

df_st_lat = df.groupby('Location')['LAT'].mean()
df_st_lon = df.groupby('Location')['LON'].mean()

result_df = pd.concat([df_st_lat, df_st_lon], axis=1)

fig, ax = plot_mapbase_diff(extent)
mappable = ax.scatter(result_df['LON'], result_df['LAT'],
                      c='black',
                      edgecolor='white',
                      s=200,
                      transform=ccrs.PlateCarree())

fig.patch.set_facecolor('none')
fig.patch.set_alpha(0.0)

# Annotate each point with its corresponding index name
for index, row in result_df.iterrows():
    ax.annotate(index, xy=(row['LON'], row['LAT']),
                xytext=(5, 5), textcoords='offset points',
                color='black', fontsize=13)

# Set the background color of the entire figure to be transparent
fig.patch.set_facecolor('none')
fig.patch.set_alpha(0.0)


plt.savefig('../Fellowships/IRL_GRF/IRL_2023/Plots/Stations_lab.png', format='png', transparent=True)
```

```python
# Create a scatter plot on the existing map
fig, ax = plot_mapbase(extent)

# Define shapes for each 'COLLECTION_TYPE'
shapes = {'OW': 'd', 'WC_B': '^', 'WC_S': 'v', 'PON': 'p', 'PW': '8', 'CORE': 'o'}

# Define custom legend labels for each 'COLLECTION_TYPE'
custom_labels = {'OW': 'Overlying Water', 'WC_B': 'Water column bottom', 'WC_S': 'Water column surface',
                 'PON': 'Poner Grab', 'PW': 'Pore Water: Core', 'CORE': 'Pore Water: Rhizon'}

# Set marker size for 35% occupancy
marker_size = 150

# Iterate over unique values of 'COLLECTION_TYPE' and plot the points
for collection_type, shape in shapes.items():
    if 'WC' in collection_type:
        # If 'WC', plot 'B' and 'S' with different shapes
        subset = df[(df['COLLECTION_TYPE'] == 'WC') & (df['Depth'].isin(['B', 'S']))]
        mappable = ax.scatter(subset['LON'], subset['LAT'], marker=shape,
                              c=subset['acdom_440_bc'],
                              edgecolor='k',
                              vmin=df['acdom_440_bc'].min(),
                              vmax=df['acdom_440_bc'].max(),
                              cmap=cmo.deep,
                              s=marker_size,
                              alpha=0.25,  # Set transparency
                              transform=ccrs.PlateCarree(),
                              label=custom_labels.get(collection_type, f'{collection_type} - {shape}'))
    else:
        # For other 'COLLECTION_TYPE', plot with the specified shape
        subset = df[df['COLLECTION_TYPE'] != 'WC']
        mappable = ax.scatter(subset['LON'], subset['LAT'], marker=shape,
                              c=subset['acdom_440_bc'],
                              edgecolor='k',
                              vmin=df['acdom_440_bc'].min(),
                              vmax=df['acdom_440_bc'].max(),
                              cmap=cmo.deep,
                              s=marker_size,
                              alpha=0.5,  # Set transparency
                              transform=ccrs.PlateCarree(),
                              label=custom_labels.get(collection_type, f'{collection_type} - {shape}'))

# Set labels and legend
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
legend = ax.legend(markerscale=0.7, bbox_to_anchor=(1.05, 1), loc='upper left', title='Sample Types')
#ax.add_artist(legend)  # Add this line to keep the colorbar from covering the legend
for lh in legend.legendHandles:
    lh.set_alpha(1.0) 

# Add colorbar below the legend
cax = fig.add_axes([.65, .3, .02, .4])  # Adjust these values as needed
cb = fig.colorbar(mappable, cax=cax, orientation='vertical')
cb.solids.set_edgecolor("face")
ax.text(-79.8, 27.9, r'$a_{\mathrm{CDOM}}$ (m$^{-1}$) 440 (nm)', size=13, rotation=90)

# Set the background color of the entire figure to be transparent
fig.patch.set_facecolor('none')
fig.patch.set_alpha(0.0)

plt.savefig('../Fellowships/IRL_GRF/IRL_2023/Plots/CDOM_map.png', format='png', transparent=True)


# Show the plot
plt.show()
```

```python
# Create a scatter plot on the existing map
fig, ax = plot_mapbase(extent)

# Define shapes for each 'COLLECTION_TYPE'
shapes = {'OW': 'd', 'WC_B': '^', 'WC_S': 'v', 'PON': 'p', 'PW': '8', 'CORE': 'o'}

# Define custom legend labels for each 'COLLECTION_TYPE'
custom_labels = {'OW': 'Overlying Water', 'WC_B': 'Water column bottom', 'WC_S': 'Water column surface',
                 'PON': 'Poner Grab', 'PW': 'Pore Water: Core', 'CORE': 'Pore Water: Rhizon'}

# Set marker size for 35% occupancy
marker_size = 150

# Iterate over unique values of 'COLLECTION_TYPE' and plot the points
for collection_type, shape in shapes.items():
    if 'WC' in collection_type:
        # If 'WC', plot 'B' and 'S' with different shapes
        subset = df[(df['COLLECTION_TYPE'] == 'WC') & (df['Depth'].isin(['B', 'S']))]
        mappable = ax.scatter(subset['LON'], subset['LAT'], marker=shape,
                              c=subset['FeT'],
                              edgecolor='k',
                              vmin=subset['FeT'].min(),
                              vmax=subset['FeT'].max(),
                              cmap=cmo.matter,
                              s=marker_size,
                              alpha=0.25,  # Set transparency
                              transform=ccrs.PlateCarree(),
                              label=custom_labels.get(collection_type, f'{collection_type} - {shape}'))
    else:
        # For other 'COLLECTION_TYPE', plot with the specified shape
        subset = df[df['COLLECTION_TYPE'] != 'WC']
        mappable = ax.scatter(subset['LON'], subset['LAT'], marker=shape,
                              c=subset['FeT'],
                              edgecolor='k',
                              vmin=subset['FeT'].min(),
                              vmax=subset['FeT'].max(),
                              cmap=cmo.matter,
                              s=marker_size,
                              alpha=0.5,  # Set transparency
                              transform=ccrs.PlateCarree(),
                              label=custom_labels.get(collection_type, f'{collection_type} - {shape}'))

# Set labels and legend
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
legend = ax.legend(markerscale=0.7, bbox_to_anchor=(1.05, 1), loc='upper left', title='Sample Types')
#ax.add_artist(legend)  # Add this line to keep the colorbar from covering the legend
for lh in legend.legendHandles:
    lh.set_alpha(1.0) 

# Add colorbar below the legend
cax = fig.add_axes([.65, .3, .02, .4])  # Adjust these values as needed
cb = fig.colorbar(mappable, cax=cax, orientation='vertical')
cb.solids.set_edgecolor("face")
ax.text(-79.8, 27.9, 'Total Dissolved Iron', size=13, rotation=90)

# Set the background color of the entire figure to be transparent
fig.patch.set_facecolor('none')
fig.patch.set_alpha(0.0)

plt.savefig('../Fellowships/IRL_GRF/IRL_2023/Plots/FeT_map.png', format='png', transparent=True)


# Show the plot
plt.show()
```
