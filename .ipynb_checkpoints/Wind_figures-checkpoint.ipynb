```python
df_42091_2021 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/42091_2021.txt', delimiter='\s+', engine='python')
df_42091_2022 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/42091_2022.txt', delimiter='\s+', engine='python')
df_LOPL1_2021 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/LOPL1_2021.txt', delimiter='\s+', engine='python')
df_LOPL1_2022 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/LOPL1_2022.txt', delimiter='\s+', engine='python')
df_PSTL1_2021 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/PSTL1_2021.txt', delimiter='\s+', engine='python')
df_PSTL1_2022 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/PSTL1_2022.txt', delimiter='\s+', engine='python')
df_BURL1_2021 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/BURL1_2021.txt', delimiter='\s+', engine='python')
df_BURL1_2022 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/BURL1_2022.txt', delimiter='\s+', engine='python')

df_PILL1_2021 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/PILL1_2021.txt', delimiter='\s+', engine='python')
df_PILL1_2022 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/PILL1_2022.txt', delimiter='\s+', engine='python')

df_TXPT2_2021 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/TXPT2_2021.txt', delimiter='\s+', engine='python')
df_TXPT2_2022 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/TXPT2_2022.txt', delimiter='\s+', engine='python')

df_EINL1_2021 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/EINL1_2021.txt', delimiter='\s+', engine='python')
df_EINL1_2022 = pd.read_csv('../Dissertation_projects/Data/NOAA_weather_data/EINL1_2022.txt', delimiter='\s+', engine='python')


df_42091_2021 = df_42091_2021.drop(index=0)
df_42091_2022 = df_42091_2022.drop(index=0)

df_LOPL1_2021 = df_LOPL1_2021.drop(index=0)
df_LOPL1_2022 = df_LOPL1_2022.drop(index=0)

df_PSTL1_2021 = df_PSTL1_2021.drop(index=0)
df_PSTL1_2022 = df_PSTL1_2022.drop(index=0)

df_BURL1_2021 = df_BURL1_2021.drop(index=0)
df_BURL1_2022 = df_BURL1_2022.drop(index=0)

df_PILL1_2021 = df_PILL1_2021.drop(index=0)
df_PILL1_2022 = df_PILL1_2022.drop(index=0)

df_TXPT2_2021 = df_TXPT2_2021.drop(index=0)
df_TXPT2_2022 = df_TXPT2_2022.drop(index=0)

df_EINL1_2021 = df_EINL1_2021.drop(index=0)
df_EINL1_2022 = df_EINL1_2022.drop(index=0)

def combine_datetime(df, year_col='#YY', month_col='MM', day_col='DD', hour_col='hh', minute_col='mm'):
    """
    Combines year, month, day, hour, and minute columns into a single datetime column.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the date and time columns.
    year_col (str): The name of the year column.
    month_col (str): The name of the month column.
    day_col (str): The name of the day column.
    hour_col (str): The name of the hour column.
    minute_col (str): The name of the minute column.

    Returns:
    pd.DataFrame: DataFrame with a new 'datetime' column.
    """
    df['datetime'] = pd.to_datetime(df[[year_col, month_col, day_col, hour_col, minute_col]].rename(
        columns={year_col: 'year', month_col: 'month', day_col: 'day', hour_col: 'hour', minute_col: 'minute'}))
    
    # Optionally drop the original columns
    df.drop(columns=[year_col, month_col, day_col, hour_col, minute_col], inplace=True)
    
    return df

# Example usage:
df_42091_2021 = combine_datetime(df_42091_2021)
df_42091_2022 = combine_datetime(df_42091_2022)

df_LOPL1_2021 = combine_datetime(df_LOPL1_2021)
df_LOPL1_2022 = combine_datetime(df_LOPL1_2022)

df_PSTL1_2021 = combine_datetime(df_PSTL1_2021)
df_PSTL1_2022 = combine_datetime(df_PSTL1_2022)

df_BURL1_2021 = combine_datetime(df_BURL1_2021)
df_BURL1_2022 = combine_datetime(df_BURL1_2022)

df_PILL1_2021 = combine_datetime(df_PILL1_2021)
df_PILL1_2022 = combine_datetime(df_PILL1_2022)

df_TXPT2_2021 = combine_datetime(df_TXPT2_2021)
df_TXPT2_2022 = combine_datetime(df_TXPT2_2022)

df_EINL1_2021 = combine_datetime(df_EINL1_2021)
df_EINL1_2022 = combine_datetime(df_EINL1_2022)


df_LOPL1 = pd.concat([df_LOPL1_2021, df_LOPL1_2022], ignore_index=True)
df_42091 = pd.concat([df_42091_2021, df_42091_2021], ignore_index=True)
df_PSTL1= pd.concat([df_PSTL1_2021, df_PSTL1_2022], ignore_index=True)
df_BURL1 = pd.concat([df_BURL1_2021, df_BURL1_2022], ignore_index=True)
df_PILL1 = pd.concat([df_PILL1_2021, df_PILL1_2022], ignore_index=True)
df_TXPT2 = pd.concat([df_TXPT2_2021, df_TXPT2_2022], ignore_index=True)
df_EINL1 = pd.concat([df_EINL1_2021, df_EINL1_2022], ignore_index=True)

```

```python
def plot_wind(ds):    
    # Drop rows with NaN values if any
    ds = ds.dropna(subset=['datetime', 'WSPD'])
    ds['WSPD'] = pd.to_numeric(ds['WSPD'], errors='coerce')
    ds = ds[ds['WSPD'] <= 98]


    fig, ax1 = plt.subplots(figsize=(20, 4))
    
    # Plotting
    ax1.plot(ds['datetime'], ds['WSPD'], color='black', label='Wind Speed')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('', color='black')
    ax1.tick_params(axis='y', labelcolor='black')
    ax1.set_ylim(0,50)
    ax1.legend(loc='upper left')
    
    fig.tight_layout()  # Adjust layout to make room for labels
    plt.show()

# Call the function with your DataFrame
plot_wind(df_42091)
plot_wind(df_LOPL1)
plot_wind(df_PSTL1)
plot_wind(df_BURL1)
plot_wind(df_PILL1)
plot_wind(df_TXPT2)
plot_wind(df_EINL1)
```

```python
def plot_winds_with_avg(df,titel,save_name):

    start_date = pd.to_datetime('2021-06-01 00:00:00')
    end_date = pd.to_datetime('2022-8-31 23:00:00')
    ds = df
    ds = ds.dropna(subset=['datetime', 'WSPD'])
    ds['WSPD'] = pd.to_numeric(ds['WSPD'], errors='coerce')
    ds.index = ds['datetime']
    ds = ds[ds['WSPD'] <= 98]
    
    wspd = ds[['WSPD']].copy()
    wspd = wspd.dropna()
    
    daily_mean = wspd.resample('D').mean()
    ten_day_mean = wspd.resample('5D').mean()
    
    # Plotting
    fig, ax1 = plt.subplots(figsize=(15, 2))

    # Plot the original wind speed data
    ax1.plot(wspd.index, wspd['WSPD'], color='lightblue', lw =1)

    # Plot the daily mean
    ax1.plot(daily_mean.index, daily_mean['WSPD'], color='black', lw =2)

    # Plot the 10-day mean
    ax1.plot(ten_day_mean.index, ten_day_mean['WSPD'], color='red', lw =1.5)

    # Set axis labels and title
    ax1.set_xlabel('')
    ax1.set_ylabel('Wind Speed (m/s)', color='black')
    ax1.tick_params(axis='y', labelcolor='black')
    ax1.set_ylim(0, 25)    
    ax1.set_xlim(start_date, end_date)

    # Add vertical lines
    ax1.axvline(x=pd.to_datetime('2021-07-15'), color='grey', linestyle='--')
    ax1.axvline(x=pd.to_datetime('2021-07-27'), color='grey', linestyle='--') 
    ax1.axvline(x=pd.to_datetime('2021-10-22'), color='grey', linestyle='--')
    ax1.axvline(x=pd.to_datetime('2021-11-14'), color='grey', linestyle='--')
    ax1.axvline(x=pd.to_datetime('2022-04-06'), color='grey', linestyle='--')
    ax1.axvline(x=pd.to_datetime('2022-04-24'), color='grey', linestyle='--')
    ax1.axvline(x=pd.to_datetime('2022-07-06'), color='grey', linestyle='--')
    ax1.axvline(x=pd.to_datetime('2022-07-28'), color='grey', linestyle='--')

    # Add shaded areas
    ax1.axvspan(pd.to_datetime('2021-07-15'), pd.to_datetime('2021-07-27'), facecolor='lightgrey', alpha=0.5)
    ax1.axvspan(pd.to_datetime('2021-10-22'), pd.to_datetime('2021-11-14'), facecolor='lightgrey', alpha=0.5)
    ax1.axvspan(pd.to_datetime('2022-04-06'), pd.to_datetime('2022-04-24'), facecolor='lightgrey', alpha=0.5)
    ax1.axvspan(pd.to_datetime('2022-07-06'), pd.to_datetime('2022-07-28'), facecolor='lightgrey', alpha=0.5)

    # Set title and adjust layout
    plt.title(titel)
    fig.tight_layout()  # Adjust layout to make room for labels
    
    plt.savefig(f'../Dissertation_projects/Figures/{save_name}.png')

    # Show plot
    
    plt.show()

df = df_BURL1
titel = 'NOAA Station BURL1 - Southwest Pass, LA (28°54\'22" N 89°25\'43" W)'
save_name = 'USGS_Station_BURL1_wind'
plot_winds_with_avg(df,titel,save_name)
```
